```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('sorting_results1.csv')  # Replace 'your_data.csv' with the path to your CSV file

# Check the first few rows of the DataFrame to verify data
print(data.head())

# Assuming the CSV has columns 'Size' and 'Comparisons'
# Customize the labels if the columns are different
sizes = data['Size']
comparisons = data['Key Comparisons']

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=sizes, y=comparisons, marker="o")

# Customize the plot with titles and labels
plt.title('Size vs Comparisons', fontsize=16)
plt.xlabel('Array Size', fontsize=14)
plt.ylabel('Number of Comparisons', fontsize=14)

# Display the graph
plt.show()
```
