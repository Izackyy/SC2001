```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('C:\Users\isaac\OneDrive\Documents\Work\SC mods\SC2001\Lab\SC2001\Project 1\data\sorting_results1.csv')  # Replace 'your_data.csv' with the path to your CSV file

# Check the first few rows of the DataFrame to verify data
print(data.head())

# Assuming the CSV has columns 'Size' and 'Comparisons'
# Customize the labels if the columns are different
sizes = data['Size']
comparisons = data['Key Comparisons']

# Calculate the second line: n * log2(n / S) + (n / 4) * S^2
theoretical_comparisons = [n * math.log2(n / 50) + (n / 4) * 50 for n in sizes]

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the actual "Size vs Comparisons"
sns.lineplot(x=sizes, y=comparisons, marker="o", label="Actual Comparisons")

# Plot the theoretical comparisons using the formula
sns.lineplot(x=sizes, y=theoretical_comparisons, marker="o", label="Theoretical Comparisons (n * log2(n / S) + (n / 4)*S)")

# Customize the plot with titles and labels
plt.title('Size vs Comparisons (Actual vs Theoretical)', fontsize=16)
plt.xlabel('Array Size', fontsize=14)
plt.ylabel('Number of Comparisons', fontsize=14)

# Display the graph with both lines
plt.legend()
plt.show()
```

```python
# Read the CSV file into a pandas DataFrame
data = pd.read_csv('sorting_results_fixed_size.csv')  # Replace 'your_data.csv' with the path to your CSV file

# Check the first few rows of the DataFrame to verify data
print(data.head())

# Assuming the CSV has columns 'Size' and 'Comparisons'
# Customize the labels if the columns are different
sizes = data['Threshold']
comparisons = data['Key Comparisons']


# Calculate the second line: n * log2(n / S) + (n / 4) * S^2
theoretical_comparisons = [1000000 * math.log2(1000000 / S) + (1000000 / 4) * S for S in sizes]

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=sizes, y=comparisons, marker="o", label = "Empirical")

# Plot the theoretical comparisons using the formula
sns.lineplot(x=sizes, y=theoretical_comparisons, marker="o", label="Theoretical Comparisons (n * log2(n / S) + (n / 4) * S)")

# Customize the plot with titles and labels
plt.title('Key Comparisons vs S', fontsize=16)
plt.xlabel('S', fontsize=14)
plt.ylabel('Number of Comparisons', fontsize=14)

# Display the graph
plt.show()
```
