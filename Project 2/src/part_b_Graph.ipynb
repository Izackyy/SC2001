```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the data from the CSV file
data = pd.read_csv('part_b_fixedV.csv')

# Step 2: Plot x against y
plt.figure(figsize=(10, 6))
plt.plot(data['|E|'], data['Key Comparisons'], marker='o', label='Empirical', color='blue')

# plot theoretical
v=1000000
theoretical = v+data['|E|']*np.log(v)
plt.plot(data['|E|'], theoretical, marker='o', label='Theoretical', color='orange')

# Step 3: Add titles and labels
plt.title('Plot of E against Key Comparisons')
plt.xlabel('E')
plt.ylabel('Key Comparisons')
plt.grid()

# Step 4: Show the plot
plt.legend()
plt.tight_layout()  # Adjust layout for better visualization
plt.show()

```

```python
# Step 1: Load the data from the CSV file
data = pd.read_csv('part_b_fixedE.csv')

# Step 2: Plot x against y
plt.figure(figsize=(10, 6))
plt.plot(data['|V|'], data['Key Comparisons'], marker='o', label='Empirical', color='blue')

# plot theoretical
E=500000
theoretical = data['|V|'] + E*np.log(data['|V|'])
plt.plot(data['|V|'], theoretical, marker='o', label='Theoretical', color='orange')

# Step 3: Add titles and labels
plt.title('Plot of V against Key Comparisons')
plt.xlabel('V')
plt.ylabel('Key Comparisons')
plt.grid()

# Step 4: Show the plot
plt.legend()
plt.tight_layout()  # Adjust layout for better visualization
plt.show()

```
